```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ROPSTEN-DATA.csv')
gp_lat = {}

for _ , row in df.iterrows():
    gasprice = row[2]
    latency = row[3]
    
    if gasprice not in gp_lat: 
        gp_lat[gasprice] = []
        
    gp_lat[gasprice].append(latency)
    
print(gp_lat)
```
