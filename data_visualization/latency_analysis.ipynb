```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ROPSTEN-DATA.csv')
gp_lat = {}

for _ , row in df.iterrows():
    gasprice = row[2]
    latency = row[3]
    
    if gasprice not in gp_lat: 
        gp_lat[gasprice] = []
        
    gp_lat[gasprice].append(latency)
    
for key, val in gp_lat.items():
    print(key, val)
```

```python
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)
ax.boxplot(gp_lat.values())
ax.set_xticklabels(gp_lat.keys())
```
