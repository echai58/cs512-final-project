```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_latency(csv_file, graph_title):
    df = pd.read_csv(csv_file)
    gp_lat = {}

    for _ , row in df.iterrows():
        gasprice = row[2]
        latency = row[3]

        if gasprice not in gp_lat: 
            gp_lat[gasprice] = []

        gp_lat[gasprice].append(latency)
        
    fig = plt.figure(figsize=(20, 10))
    ax = fig.add_subplot()
    ax.set_title(graph_title)
    ax.set_xlabel('gas price (Gwei)')
    ax.set_ylabel('latency (seconds)')
    ax.boxplot(gp_lat.values(), showfliers=False)
    ax.set_xticklabels(gp_lat.keys())
    fig.savefig(f'{graph_title}.jpg', bbox_inches='tight')
```

```python
plot_latency('ROPSTEN-DATA.csv', 'Ropsten Transaction Latency')
```

```python
plot_latency('GOERLI-DATA.csv', 'Goerli Transaction Latency')
```
